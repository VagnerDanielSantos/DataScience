# Ordenando DataFrames

```python
import pandas
import numpy
```

```python
colunas_filmes = ['star_rating', 'title', 'content_rating', 'genre', 'duration', 'actors_list']

file = 'E:\\Udemy - Cursos\\Data Science\\Pacotes\\movies.csv'
filmes = pandas.read_csv(file, header = 0, names = colunas_filmes)
filmes.columns
```

```python
filmes.head()
```

```python
# Ordenando por título

filmes.title.sort_values().head()
```

```python
# O retorno é uma Serie

A = filmes.title.sort_values()

type(A)
```

```python
A
```

```python
# Ordem Descendente

filmes.title.sort_values(ascending = False).head(10)
```

```python
# Há outra forma de ordenar um DataFrame, desta forma, o retorno é um DataFrame

filmes.sort_values('title').head(10)
```

```python
# Podemos também, ordenar de forma descendente

filmes.sort_values('title', ascending = False).head(10)
```

```python
# Apenas ordenar, não altera o DataFrame original

filmes.head()
```

```python
# Ordenando por classificação e duração

filmes.sort_values([ 'genre', 'duration']).head(10)
```

# Filtrando um DataFrame

```python
# Criando uma lista para saber os filmes com uma duração maior do que 150 minutos

boleanos = []

for length in filmes.duration:
    if length >= 150:
        boleanos.append(True)
    else:
           boleanos.append(False)
```

```python
boleanos[: 20]
```

```python
type(boleanos)
```

```python
# Verificando o tamanho da lista

len(boleanos)
```

# É preciso converter a Lista em uma Serie para poder aplicar como filtro de DataFrame PANDAS

```python
# Convertendo a Lista em Serie

duracao = pandas.Series(boleanos)

duracao.head(20)
```

```python
type(duracao)
```

```python
# É possível usar a notação de colchetes para indexar ou para aplicar filtros de series boleanos

print(filmes['title'] [: 15])
```

```python
filmes[duracao]
```

```python
# É possível ainda, continuar a notação trazendo somente as colunas desejadas

filmes[duracao] [ ['title', 'duration', 'genre'] ]
```

# É possível simplificar o filtro conforme abaixo ou ainda escrevendo em List Comprehension

```python
# Forma simplificada

A = filmes.duration >= 180

A.head(20)
```

```python
filmes[A].head(16)
```

```python
# Escrevendo por List Comprehension

filmes[filmes.duration >= 180].head(20)
```

```python
# É possível também retornar outra série com a coluna que desejamos

genero_maior_200 = filmes[filmes.duration >= 200].genre
```

```python
genero_maior_200
```

```python
type(genero_maior_200)
```

```python
# É possível utilizar a notação de colchetes para se fazer a mesma coisa

genero_maior_200 = filmes[filmes.duration >= 200] ['genre']
```

```python
genero_maior_200
```

```python
# Porém, por uma boa prática, utilizamos sempre o método LOC que nada mais é do que definir uma condição e filtrar por labels 
# Veremos o ILOCcom mais detalhes 

filmes.loc[filmes.duration >= 200, 'genre']
```
